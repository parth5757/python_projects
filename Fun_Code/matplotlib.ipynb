```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, timedelta

# Assuming the investment dates are stored in a dictionary
# For example: {"Bajaj Finances": "2024-01-15", "Avenue Supermarts": "2024-02-20",...}
investment_dates = {
    "Bajaj Finances": "2024-01-15",
    "Avenue Supermarts": "2024-02-20",
    "Polycab India": "2024-03-25",
    "Asian Paints": "2024-04-30",
    "Tata Elexi": "2024-05-05",
    "L&T": "2024-06-10",
    "Titan": "2024-07-15"
}

# Get the current date
current_date = datetime.now()

# Calculate total time horizon in days for each stock
total_time_horizons_days = {}
for label, investment_date_str in investment_dates.items():
    investment_date = datetime.strptime(investment_date_str, "%Y-%m-%d")
    total_time_horizons_days[label] = (current_date - investment_date).days

# First pie chart: invested amount
y = np.array([33896.15, 16293.64, 7531.66, 8876.01, 5898.00, 4681.70, 5104.26])
mylabels = ["Bajaj Finances", "Avenue Supermarts", "Polycab India", "Asian Paints", "Tata Elexi", "L&T", "Titan"]
plt.pie(y, labels=mylabels, startangle=90, autopct="%1.1f%%")
plt.title("Invested Amount")
plt.show()

print()

# Second pie chart: current value
x = np.array([33593.75, 18774.80, 12861.80, 8637.75, 7069.00, 6587.70, 6507.30])
mylabels1 = ["Bajaj Finances", "Avenue Supermarts", "Polycab India", "Asian Paints", "Tata Elexi", "L&T", "Titan"]
plt.pie(x, labels=mylabels1, startangle=90, autopct="%1.1f%%")
plt.title("Current Value")
plt.show()

# Calculate and print the difference, daily return rate, and total time horizon for each stock
for label, total_time_horizon_days in total_time_horizons_days.items():
    difference = (x[mylabels.index(label)] - y[mylabels.index(label)]) / y[mylabels.index(label)] * 100
    daily_return_rate = ((x[mylabels.index(label)] - y[mylabels.index(label)]) / y[mylabels.index(label)]) / total_time_horizon_days * 100
    print(f"{label}: Difference = {difference:.2f}%, Daily Return Rate = {daily_return_rate:.2f}%, Total Time Horizon = {total_time_horizon_days} days")
```
